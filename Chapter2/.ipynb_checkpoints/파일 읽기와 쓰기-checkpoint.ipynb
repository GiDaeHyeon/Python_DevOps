#### 4

```python
file_path = './untitled.json'
open_file = open(file_path, 'r')
text = open_file.read()
print(text)
print(len(text))
print(type(text))
open_file.close()
```

```python
import json
from pprint import pprint

with open(file_path, 'r') as opened_file:
    dic = json.load(opened_file)
pprint(dic)
print(type(dic))
```

```python
with open(file_path, 'w') as opened_file:
    dic["Hello"] = "World"
    dic = json.dump(dic, opened_file)
    
with open(file_path, 'r') as opened_file:
    dic = json.load(opened_file)
    
pprint(dic)
```
